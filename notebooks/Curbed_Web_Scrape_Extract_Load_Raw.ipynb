```python
from firecrawl import FirecrawlApp
import pandas as pd
from sqlalchemy import create_engine
```

```python
app = FirecrawlApp(api_key='fc-bd419862f2134069b9263f8ebd36bd1a')
```

```python
scrape_result = app.scrape_url(
    url,
    formats=['json'],
    jsonOptions={
        'prompt': 'List the major Apple Store locations mentioned in the article.'
    }
)
```

```python

store = scrape_result.json

print(store)
```
